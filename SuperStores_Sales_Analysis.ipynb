# Time Series

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.quarter
```

```python
monthly_sales = df.groupby(pd.Grouper(key='Order Date', freq='M'))['Sales'].sum()

monthly_sales.plot(figsize=(12,5))
plt.title("Monthly Sales Trend")
plt.show()
```

```python
monthly_sales.rolling(3).mean().plot()
```

### This is used to calculate and plot a 3-month moving average,very common in sales analysis and time-series projects.

# Skewness & Kurtosis

```python
df['Sales'].skew()
```

```python
df['Sales'].kurt()
```

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['Sales'] < Q1 - 1.5*IQR) | (df['Sales'] > Q3 + 1.5*IQR)]
```

```python
sns.boxplot(x=df['Sales'])
plt.title("Sales Outlier Detection")
plt.show()
```
